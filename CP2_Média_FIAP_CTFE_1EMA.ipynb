```
def calcularmediacps():
  cps = int(input("Quantas CP's Tem?: "))
  if cps >= 3:
    notas_cp = []

    for i in range(cps):
      CP_1 = float(input(f"Qual a nota da CP {i + 1}?: "))
      notas_cp.append(CP_1)

    notas_cp.sort(reverse=True)
    media_cp = (sum(notas_cp[:2])/2) * 0.2
    return media_cp
  else:
    return "Coloque no minimo 3 CPS"

calcularmediacps()
```

```
def calcularmediasprint():
  print("---------------------------")
  spring1 = float(input("Insira o valor da 1° Sprint: "))
  spring2 = float(input("Insira o valor da 2° Sprint: "))
  print("---------------------------")

  spring = ((spring1 + spring2) / 2) * 0.2

  return spring
```

```
def calcularmediafinal():

  media_cp = calcularmediacps()
  media_spring = calcularmediasprint()
  media_final = media_cp + media_spring
  if media_final >= 6:
      print("---------------------------")
      print(f"Sua média parcial é {media_final:.2f}, você já atingiu o mínimo!")
      print("---------------------------")
  else:
      print("---------------------------")
      print(f"Sua média parcial é {media_final:.2f}, faltam {6 - media_final:.2f} pontos")
      print("---------------------------")

  mediaglobal = (float(input("Insira o valor da Global Solution: "))) * 0.6
  print("---------------------------")

  media_final = media_cp + mediaglobal + media_spring

  return media_final
```

```
print("---------------------------")
print("1°Semetre")
print("---------------------------")
media1 = calcularmediafinal() * 0.4
print("---------------------------")
print("2°Semetre")
print("---------------------------")
media2 = calcularmediafinal() * 0.6
media_final = media1 + media2

if media_final >= 6:
  print("---------------------------")
  print("Aprovado")
  print("---------------------------")
elif media_final >= 4:
  print("---------------------------")
  print("Exame")
  print("---------------------------")
  nota_exame = float(input("Insira a nota do exame: "))
  if ((media_final + nota_exame) / 2) >= 6:
    print("---------------------------")
    print("Aprovado")
    print("---------------------------")
  else:
    print("---------------------------")
    print("Reprovado")
    print("---------------------------")
else:
  print("---------------------------")
  print("Reprovado")
  print("---------------------------")
```
