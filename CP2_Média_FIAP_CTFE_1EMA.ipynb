```
def calcularmediacps():
  cps = int(input("Quantas CP's Tem?: "))
  if cps >= 3:
    notas_cp = []

    for i in range(cps):
      CP_1 = float(input(f"Qual a nota da CP {i + 1}?: "))
      notas_cp.append(CP_1)

    notas_cp.sort(reverse=True)
    media_cp = (sum(notas_cp[:2])/2) * 0.2
    return media_cp
  else:
    return "Coloque no minimo 3 CPS"
```

```
def calcularmediasprint():
  spring1 = float(input("Insira o valor da 1° Sprint: "))
  spring2 = float(input("Insira o valor da 2° Sprint: "))

  spring = ((spring1 + spring2) / 2) * 0.2

  return spring
```

```
def calcularmediafinal():
  mediaglobal = (float(input("Insira o valor da Global Solution: "))) * 0.6
  media_cp = calcularmediacps()
  media_spring = calcularmediasprint()

  media_final = media_cp + mediaglobal + media_spring

  return media_final
```

```
media_final = calcularmediafinal()

if media_final >= 6:
  print("Aprovado")
elif media_final >= 4:
  print("Exame")
  nota_exame = float(input("Insira a nota do exame: "))
  if ((media_final + nota_exame) / 2) >= 6:
    print("Aprovado")
  else:
    print("Reprovado")
else:
  print("Reprovado")
```
